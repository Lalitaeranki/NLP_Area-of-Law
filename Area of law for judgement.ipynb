# NLP Approach to identify Area of Law for Indian Law Judgement


```python
#making the dataframe from csv and text files
import pandas as pd
judgement_file=pd.read_csv("Interview_Mapping.csv")
```

```python
#List of Judgements from the Judgement Dataframe
file_list=judgement_file.iloc[:,0].tolist()
file_list
```

```python
#creating dataframe for text
text_doc=[]
for name in file_list:
     with open("Fixed Judgements/" + name + ".txt",'r', encoding='utf-8',errors='ignore') as file:
            lines=file.read().splitlines()
            text_doc.append({
                       "Judgements":name,
                       "Text":lines }) 
            
          
```

```python
text_df= pd.DataFrame.from_dict(text_doc)
```

```python
text_df.head()
```

```python
#merging the dataframe on Judgement column
Combined_data=pd.merge(judgement_file,text_df,how="left",on="Judgements")
```

```python
Combined_data.head()
```

```python
Combined_data.rename(columns={"Area.of.Law":"Area_of_Law"},inplace=True)
```

```python
Combined_data['Text_new'] = Combined_data['Text'].apply(lambda x: ' '.join(x))
```

```python
Combined_data.head(10)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
#Converting to lower case
Combined_data['Text_new'] = Combined_data['Text_new'].str.lower()
```

```python
#Converting each entry to list token(Tokennizing)
def identify_tokens(row):
    Text_new = row['Text_new']
    tokens = nltk.word_tokenize(Text_new)
    # taken only words (not punctuation)
    token_words = [w for w in tokens if w.isalpha()]
    return token_words

Combined_data['token_words'] = Combined_data.apply(identify_tokens, axis=1)
```

```python
#Stemming :stemming to root word

from nltk.stem import PorterStemmer
stemming = PorterStemmer()

```

```python
def stem_list(row):
    my_list = row['token_words']
    stemmed_list = [stemming.stem(word) for word in my_list]
    return (stemmed_list)

Combined_data['stemmed_words'] = Combined_data.apply(stem_list, axis=1)
```

```python
Combined_data.head()
```

```python
#Removing StopWords
from nltk.corpus import stopwords
stops = set(stopwords.words("english"))                  

def remove_stops(row):
    my_list = row['stemmed_words']
    meaningful_words = [w for w in my_list if not w in stops]
    return (meaningful_words)

Combined_data['stem_meaningful'] = Combined_data.apply(remove_stops, axis=1)
```

```python
def rejoin_words(row):
    my_list = row['stem_meaningful']
    joined_words = ( " ".join(my_list))
    return joined_words

Combined_data['Cleaned'] = Combined_data.apply(rejoin_words, axis=1)
```
